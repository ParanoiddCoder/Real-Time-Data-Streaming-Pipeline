```python
from elasticsearch import Elasticsearch, helpers
import json

# Connect to Elastic Cloud
client = Elasticsearch(
    "https://my-deployment-374dc0.es.us-east-2.aws.elastic-cloud.com",
    basic_auth=("elastic", "5Qb5TVyJcll1ZqYvjC2bZ2Vn"),
    verify_certs=False  # NOTE: Set to True in production
)

# Read and prepare data for bulk upload
actions = []

with open("/kaggle/input/yelp-dataset/yelp_academic_dataset_business.json") as f:
    for line in f:
        try:
            doc = json.loads(line.strip())
            action = {
                "_index": "yelp_deployment",
                "_source": doc
            }
            actions.append(action)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")

# Upload to Elasticsearch
if actions:
    helpers.bulk(client, actions)
    print("Bulk upload successful.")
else:
    print("No valid documents to upload.")
```
